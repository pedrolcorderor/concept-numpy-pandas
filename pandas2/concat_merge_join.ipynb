### concat()

para concatener recordar usar [], y el ignore_index es para que respete y reorganice los indices y por defecto lo hace por el axis 0 (flows)

```python
pd.concat([df1,df2],ignore_index=True)
```

cunado no tiene el ignore_index(ver los indices)

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis=1)
```

### merge()

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq
```

```python
der
```

```python
izq.merge(der, on='key')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2','k3'],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

```python
izq = pd.DataFrame({'key':['k0','k1','k2','k3'],
 'A':['A0','A1','A2','A3'],
 'B':['B0','B1','B2','B3']})

der = pd.DataFrame({'key_2':['k0','k1','k2',np.nan],
 'C':['C0','C1','C2','C3'],
 'D':['D0','D1','D2','D3']})
```

en caso de que una de las llaves tenga un valor en NaN, al hacer merge no lo va a traer porque no coinciden los dos .

```python
der
```

```python
izq.merge(der, left_on='key', right_on='key_2')
```

pero si queremos esos valores que tiene el de la izquierda o derecha apesar que el de la derecha no tenga usamos how='right'  (el valor por defecto es el inner)


### how

```python
izq.merge(der, left_on='key', right_on='key_2', how='left')
```

```python
izq.merge(der, left_on='key', right_on='key_2', how='right')
```

### join() - Index match

¿Cuál es la principal diferencia entre merge y join en pandas?

Merge puede unificar df en base a los índices o a los valores de las columnas.
Join solo lo puede hacer en base a los índices. Es una caso específico de merge.

```python
izq = pd.DataFrame({'A':['A0','A1','A2'],
 'B':['B0','B1','B2']},
 index=['k0','k1','k2'])

der = pd.DataFrame({'C':['C0','C1','C2'],
 'D':['D0','D1','D2']},
 index=['k0','k2','k3'])
```

el how='outer' en join viene por defecto o se lo podemos colocar

```python
izq.join(der,how='outer')
```

join hace una union directamente desde los indices

y podemos pedir solamente los que los indices hacen macth 

```python
izq.join(der,how='inner')
```

```python
izq.join(der,how='left')
```

```python
izq.join(der,how='right')
```
